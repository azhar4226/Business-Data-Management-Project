```python
# Data Preprocessing
sales_data['Price'] = pd.to_numeric(sales_data['Price'], errors='coerce')
sales_data['Date'] = pd.to_datetime(sales_data['Date'], errors='coerce')
purchase_data['Date'] = pd.to_datetime(purchase_data['Date'], errors='coerce')
purchase_data['Cost per Unit'] = pd.to_numeric(purchase_data['Cost per Unit'], errors='coerce')
purchase_data['Units'] = pd.to_numeric(purchase_data['Units'], errors='coerce')
```

```python
# Drop missing or irrelevant rows
sales_data = sales_data.dropna(subset=['Price', 'Brand', 'Model'])
purchase_data = purchase_data.dropna(subset=['Cost per Unit', 'Units'])
```

```python
# Top-Selling Products by Revenue
top_revenue_products = sales_data.groupby('Model')['Price'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
axis = sns.barplot(x=top_revenue_products.values, y=top_revenue_products.index, palette='plasma', hue=top_revenue_products.index)
plt.gca()
plt.tight_layout()
plt.xlabel('Revenue')
plt.ylabel('Model')
plt.title('Top-Selling Products by Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pareto Analysis for Revenue
# ------------------------------
pareto_data = sales_data.groupby('Model')['Price'].sum().sort_values(ascending=False)
pareto_cumulative = pareto_data.cumsum() / pareto_data.sum()

fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart for individual product revenue
sns.barplot(x=pareto_data.index[:10], y=pareto_data.values[:10], palette='coolwarm', ax=ax1)
ax1.set_title("Pareto Analysis for Revenue (Top Products)", fontsize=16)
ax1.set_xlabel("Product Model", fontsize=12)
ax1.set_ylabel("Revenue", fontsize=12)
ax1.tick_params(axis='x', rotation=45)

# Line chart for cumulative percentage
ax2 = ax1.twinx()
sns.lineplot(x=pareto_data.index[:10], y=pareto_cumulative[:10] * 100, color='orange', marker='o', ax=ax2)
ax2.set_ylabel("Cumulative Percentage (%)", fontsize=12)
ax2.axhline(80, color='red', linestyle='--', label='80% Threshold')
ax2.legend(loc='upper left')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Define price bins and labels
price_bins = [0, 5000, 10000, 15000, 20000, 25000, 30000, 35000, 40000, 50000]
price_labels = ['0-5k', '5k-10k', '10k-15k', '15k-20k', '20k-25k',
                '25k-30k', '30k-35k', '35k-40k', '40k-50k']

# Add a Price Range column to the dataset
sales_data['Price_Range'] = pd.cut(sales_data['Price'], bins=price_bins, labels=price_labels)

# Count the number of models within each price range
price_range_counts = sales_data.groupby('Price_Range')['Model'].nunique()

# Plot the histogram (bar chart)
plt.figure(figsize=(10, 6))
sns.barplot(x=price_range_counts.index, y=price_range_counts.values, palette="Blues_r")
plt.title("Number of Product Models Sold at Different Price Ranges", fontsize=14)
plt.xlabel("Price Range (₹)", fontsize=12)
plt.ylabel("Number of Product Models", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Calculate Total Units Sold per Model (using sales_data)
model_sales = sales_data.groupby('Model')['Price'].size().reset_index()
model_sales.rename(columns={'Price': 'Total_Units_Sold'}, inplace=True)

# Calculate Total Units Purchased and Average Inventory (using purchase_data)
purchase_summary = purchase_data.groupby('Model').agg({'Units': 'sum', 'Date': 'nunique'})
purchase_summary.rename(columns={'Units': 'Total_Units_Purchased', 'Date': 'Number_of_Purchases'}, inplace=True)
purchase_summary['Average_Inventory'] = purchase_summary['Total_Units_Purchased'] / purchase_summary['Number_of_Purchases']

# Merge DataFrames
inventory_turnover_data = pd.merge(model_sales, purchase_summary, on='Model', how='left')

# Calculate Inventory Turnover
inventory_turnover_data['Inventory_Turnover'] = inventory_turnover_data['Total_Units_Sold'] / inventory_turnover_data['Average_Inventory']

# Sort and visualize
turnover_data = inventory_turnover_data.sort_values(by=['Inventory_Turnover'], ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='Inventory_Turnover', y='Model', data=turnover_data.head(10), palette='Blues_r')
plt.title("Top 10 Models by Inventory Turnover", fontsize=14)
plt.xlabel("Inventory Turnover Rate", fontsize=12)
plt.ylabel("Product Model", fontsize=12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Calculate Total Units Sold per Model (using sales_data)
units_sold_by_model = sales_data.groupby('Model')['Price'].size().reset_index()
units_sold_by_model.rename(columns={'Price': 'Total_Units_Sold'}, inplace=True)

# 2. Calculate Total Units Purchased per Model (using purchase_data)
units_purchased_by_model = purchase_data.groupby('Model')['Units'].sum().reset_index()
units_purchased_by_model.rename(columns={'Units': 'Total_Units_Purchased'}, inplace=True)

# 3. Merge DataFrames
turnover_data = pd.merge(units_sold_by_model, units_purchased_by_model, on='Model', how='left')

# 4. Calculate Turnover Rate (using Total Units Sold / Total Units Purchased)
turnover_data['Turnover_Rate'] = turnover_data['Total_Units_Sold'] / turnover_data['Total_Units_Purchased']

# 6. Outlier Detection and Removal using IQR
# Q1 = turnover_data['Turnover_Rate'].quantile(0.25)
# Q3 = turnover_data['Turnover_Rate'].quantile(0.75)
# IQR = Q3 - Q1
# lower_bound = Q1 - 1.5 * IQR
# upper_bound = Q3 + 1.5 * IQR

filtered_turnover_data = turnover_data[turnover_data['Turnover_Rate'] < 1]
# 7. Sort and Visualize (using filtered data)
turnover_rate_sorted = filtered_turnover_data.sort_values(by=['Turnover_Rate'], ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Turnover_Rate', y='Model', data=turnover_rate_sorted, palette='magma')
plt.title('Top 10 High-Turnover Models (Prioritized for Restocking, Outliers Removed)')
plt.xlabel('Turnover Rate (Units Sold / Units Purchased)')
plt.ylabel('Product Model')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Aggregate sales and purchase data
quantity_sold = sales_data.groupby('Model').size().rename('Quantity_Sold')
quantity_purchased = purchase_data.groupby('Model')['Units'].sum().rename('Quantity_Purchased')

# 2. Combine into a single DataFrame
inventory_data = pd.concat([quantity_sold, quantity_purchased], axis=1).reset_index().fillna(0)

# 3. Calculate Stock Status
inventory_data['Stock_Status'] = inventory_data.apply(lambda row: 'Stock-Out' if row['Quantity_Sold'] > row['Quantity_Purchased'] else 'Overstock' if row['Quantity_Sold'] < row['Quantity_Purchased'] else 'Balanced', axis=1)


# 4. Scatter Plot for Stock-Out and Overstock Trends
plt.figure(figsize=(12, 6))
sns.scatterplot(
    x='Quantity_Purchased',
    y='Quantity_Sold',
    data=inventory_data,
    hue='Stock_Status',  # Color by stock status
    size='Quantity_Sold',
    palette={'Stock-Out': 'red', 'Overstock': 'blue', 'Balanced': 'green'},  # Custom colors
    sizes=(50, 300),
    alpha=0.7
)

# 5. Plot diagonal to separate stock-out and overstock regions
max_value = max(inventory_data['Quantity_Purchased'].max(), inventory_data['Quantity_Sold'].max())
plt.plot([0, max_value], [0, max_value], 'k--', label='Ideal Stock Balance')  # Diagonal line

# 6. Labels and Title
plt.title("Stock-Out vs Overstock Trends", fontsize=16)
plt.xlabel("Quantity Purchased (Supply)", fontsize=12)
plt.ylabel("Quantity Sold (Demand)", fontsize=12)
plt.legend(title="Stock Status", loc='upper left')  # Legend for stock status
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 1. Data Preparation
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
sales_data['Month'] = sales_data['Date'].dt.to_period('M')

# 2. Merge Sales and Purchase Data for COGS Calculation
sales_data = pd.merge(sales_data, purchase_data[['Model', 'Cost per Unit']], on='Model', how='left')

# 3. Calculate Revenue, COGS, and Gross Profit
sales_data['Revenue'] = sales_data['Price']
sales_data['COGS'] = sales_data['Cost per Unit']  # Assuming 'Cost' is the purchase cost per unit
sales_data['Gross_Profit'] = sales_data['Revenue'] - sales_data['COGS']

# 4. Aggregate by Month
monthly_data = sales_data.groupby('Month').agg({'Revenue': 'sum', 'Gross_Profit': 'sum'})

# 5. Calculate Net Profit (subtracting fixed cost)
monthly_data['Net_Profit'] = monthly_data['Gross_Profit'] - 22200  # Subtract fixed cost of 22200

# 6. Plot Net Profit Trend
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_data.index.astype(str), y=monthly_data['Net_Profit'], marker='o', color='green')
plt.title("Net Profit Trend by Month", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Net Profit (₹)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
